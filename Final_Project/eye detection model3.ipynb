```python
nipuni=cv2.imread("/Users/nipuni/Desktop/Nipuni.jpg")
```

```python
imageio.imshow(nipuni)
```

```python
rgb=cv2.cvtColor(nipuni,cv2.COLOR_BGR2RGB)
imageio.imshow(rgb)
```

```python
# Let's show Haar cascade for eye detector has false positive
eyedetect= cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye.xml")
geteye=eyedetect.detectMultiScale(rgb)
```

```python
geteye
```

```python
gozler=[]
for (x,y,w,h) in geteye:
    gozler.append(rgb[y:y+h, x:x+w])
for i in range(len(gozler)):
    plt.figure()
    imageio.imshow(gozler[i])
```

# Now let's try eye detection using landmarks



```python
import dlib
from imutils import face_utils
detector = dlib.get_frontal_face_detector()
# Coordinates of the face
detect=detector(rgb,1)
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
shape=predictor(rgb,detect[0])
print(shape)
```

```python
detect[0]
```

```python
x1=shape.part(36).x
x2=shape.part(39).x
#43 46
y1=shape.part(42).y
y2=shape.part(45).y
print("{1} : {2} {3} : {4}",x1,x2,y1,y2)
```

```python
frans=rgb[y2:y1,x1:x2]
imageio.imshow(frans)
```

```python
frans=rgb[y2-15:y1+15,x1:x2]
imageio.imshow(frans)
#39 42
#37 40
```
